## Task 3 : Scape all the links using Beauttiful Soup

```python
#block 1
def getURL(): # create a function for scraping all the links in a page
    page_source = BeautifulSoup(driver.page_source,'html.parser') # parse all the information in that page
    profiles = page_source.find_all('a',class_ = 'job_link') #find all the "a" tag with the 'job_link' class
    all_profiles_URL =[] # create empy list too contain all the links
    sleep(1)
    for i,profile in enumerate(profiles): # loop through all the links in the profile
        profile_URL = profile.get('href') # find the link using href
        if profile_URL not in all_profiles_URL: # if that link does not occur in the list, append it to the list. Otherwise remove it
            all_profiles_URL.append(profile_URL)
    return all_profiles_URL
```

### Task 3 : Write the function to scrape the data

```python
# Define scaping function
def crawl_data():
    #block 1
    def getURL(): # create a function for scraping all the links in a page
        page_source = BeautifulSoup(driver.page_source,'html.parser') # parse all the information in that page
        profiles = page_source.find_all('a',class_ = 'job_link') #find all the "a" tag with the 'job_link' class
        all_profiles_URL =[] # create empy list too contain all the links
        sleep(1)
        for i,profile in enumerate(profiles): # loop through all the links in the profile
            profile_URL = profile.get('href') # find the link using href
            if profile_URL not in all_profiles_URL: # if that link does not occur in the list, append it to the list. Otherwise remove it
                all_profiles_URL.append(profile_URL)
        return all_profiles_URL

    #block 2
    try:
        #Define the number of page you want tto scrape
        number_of_crawling_page = int(input('Please enter the number of page you want to crawl : '))
        all_URL_page =[] # create an empty list
        for i in range(number_of_crawling_page) :
            url_one_page= getURL() # get all the links in 1 page
            sleep(2) # slowdown the program
            print('Get URL done')
            Next_button = driver.find_element(By.CLASS_NAME ,'mdi-chevron-right') # move to next page
            Next_button.click() #click on the next button and continue the loop
            print('Click next done')
            all_URL_page = all_URL_page + url_one_page
            sleep(2)
            print('add to the list done')
        print('all done')
    except:
        pass
    all_URL_page

    #block 3
    import csv
    with open(r'F:\DATA ANALYST\Python\careerviet_crawl.csv', 'w', encoding = 'utf-8-sig', newline = '') as output_file:
        headers = ['Company name','Title','Domain','Position','Experience','Salary','Location']
        
        #method to write header
        writer = csv.DictWriter( output_file , delimiter = ',', lineterminator = '\n' , fieldnames = headers)
        writer.writeheader()
        #Loop through each post and scrape all data
        for link in all_URL_page: # for loop run
            try:
                driver.get(link) # go to the link
                page_source = BeautifulSoup(driver.page_source, 'html.parser') # parse all the data from that page
                #Find theh position and company name
                info_name_position_div = page_source.find('div' , class_ = 'apply-now-content') #find the div tag which contain company name , title
                position = info_name_position_div.find('h1', class_ = 'title').get_text().strip() #extract the position of the link
                company_name = info_name_position_div.find('a', class_ = 'employer job-company-name').get_text().strip() #extract the company name of the link
                #good
                
                #Find the domain
                info_domain_div = page_source.find_all('div' , class_ = 'detail-box has-background')[0] #find the div tag which contain domain of the link
                domain = info_domain_div.find_all('li')[1].find('a').get_text().strip() #extract the domain of the link
                #good
                
                #Find the salary, experience, level
                info_salary_exp_level_div = page_source.find_all('div' , class_ = 'detail-box has-background')[1] #find the div tag which contain salary, experience, level
                salary = info_salary_exp_level_div.find_all('p')[0].get_text().strip() #extract the salary of the link
                experience =" ".join(info_salary_exp_level_div.find_all('p')[1].get_text().replace('\n','').strip().split()) #extract the experience of the link
                level = info_salary_exp_level_div.find_all('p')[2].get_text().strip() #extract the level of the link
                #good
                
                #Find the location
                info_location_div = page_source.find('div' , class_ = 'detail-box')
                location = info_location_div.find('p').get_text().strip() 
                #good
                
                print(position)
                print(company_name)
                print(domain)
                print(salary)
                print(experience)
                print(level)
                print(location)
                writer.writerow({headers[0]:company_name ,headers[1]:position,headers[2]:domain,headers[3]:level, \
                headers[4]:experience , headers[5]:salary,headers[6]:location})
                sleep(1) #slowdown the program
            except:
                pass

    return None
```
